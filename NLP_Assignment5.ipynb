```python
import kagglehub
import pandas as pd
import string

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
path = kagglehub.dataset_download("sahideseker/chatbot-intent-classification-dataset")

print("Dataset downloaded at:", path)
```

```python
import os

print(os.listdir(path))

file_path = os.path.join(path, "chatbot_intent_classification.csv")
df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
```

```python
def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text


df["user_input"] = df["user_input"].apply(preprocess)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df["user_input"],
    df["intent"],
    test_size=0.2,
    random_state=42
)
```

```python
vectorizer = TfidfVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
def predict_intent(text):
    text = preprocess(text)
    vec = vectorizer.transform([text])
    return model.predict(vec)[0]
```

```python
print(predict_intent("I want to cancel my subscription"))
print(predict_intent("I forgot my password"))
print(predict_intent("What are your timings"))
```

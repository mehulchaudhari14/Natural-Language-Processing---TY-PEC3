```python
import re
import numpy as np
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from gensim.models import Word2Vec
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def preprocess_corpus(text):

    sentences = sent_tokenize(text.lower())

    processed_sentences = []
    for sent in sentences:

        clean_sent = re.sub(r'[^\w\s]', '', sent)

        tokens = word_tokenize(clean_sent)
        if tokens:
            processed_sentences.append(tokens)

    return processed_sentences
```

```python
def compute_cosine_similarity(vec_a, vec_b):

    dot_product = np.dot(vec_a, vec_b)
    norm_a = np.linalg.norm(vec_a)
    norm_b = np.linalg.norm(vec_b)

    if norm_a == 0 or norm_b == 0:
        return 0.0
    return dot_product / (norm_a * norm_b)
```

```python
def train_word_embeddings():
    print("--- Starting Word Embedding Training ---")

    corpus = """
    Artificial intelligence enables smart systems to perform tasks.
    Machine learning allows systems to learn patterns from data.
    Deep learning uses neural networks for advanced predictions.
    Natural language processing helps machines understand human language.
    Neural networks are the foundation of deep learning models.
    Supervised and unsupervised learning are types of machine learning.
    NLP is used in chatbots, translation, and sentiment analysis.
    AI applications are widely used in healthcare, finance, and automation.
    """

    D = 10
    window_size = 2
    epochs = 100

    print("Preprocessing text...")
    sentences = preprocess_corpus(corpus)

    print("Training Word2Vec model (Skip-gram)...")
    model = Word2Vec(
        sentences=sentences,
        vector_size=D,
        window=window_size,
        sg=1,
        min_count=1,
        epochs=epochs
    )

    word_vectors = model.wv
    print("\nModel trained successfully! Vocabulary size:", len(word_vectors.index_to_key))

    target_word = "learning"
    print(f"\nVector representation for '{target_word}' (Dimension {D}):")
    print(np.round(word_vectors[target_word], 3))

    print("\n--- Semantic Similarity Analysis ---")

    pairs_to_test = [
        ("artificial", "intelligence"),
        ("machine", "learning"),
        ("deep", "learning"),
        ("language", "processing"),
        ("learning", "automation")
    ]

    for w1, w2 in pairs_to_test:
        if w1 in word_vectors and w2 in word_vectors:
            vec1 = word_vectors[w1]
            vec2 = word_vectors[w2]

            similarity = compute_cosine_similarity(vec1, vec2)
            print(f"Similarity('{w1}', '{w2}') \t-> {similarity:.4f}")

    print("\n--- END ---")


if __name__ == "__main__":
    train_word_embeddings()
```

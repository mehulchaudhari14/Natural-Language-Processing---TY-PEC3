```python
import nltk
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
```

```python

paragraph="""Hello world good morning have a great day.A paragraph is a collection of sentences focused on a single main idea. Typically, effective paragraphs include aparagragh to introduce the central point, several supporting sentences for details, and a concluding sentence to summarize or transition"""
```

```python
words=nltk.word_tokenize(paragraph)

print(words)
```

```python
tokens1=nltk.sent_tokenize(paragraph)
print(tokens1)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
for word in words:
  print(word+"---->"+ps.stem(word))


```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
for word in words:
  print(word+"--->"+lemmatizer.lemmatize(word))
```

```python
import re
from collections import defaultdict
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = text.split()
    return words
```

```python
def train_ngram_model(words, N):
    ngram_counts = defaultdict(lambda: defaultdict(int))

    for i in range(len(words) - N + 1):

        context = tuple(words[i : i + N - 1])
        next_word = words[i + N - 1]
        ngram_counts[context][next_word] += 1

    ngram_probs = defaultdict(dict)

    for context, next_words in ngram_counts.items():
        total_count = sum(next_words.values())

        for next_word, count in next_words.items():
            probability = count / total_count
            ngram_probs[context][next_word] = probability

    return ngram_probs
```

```python
def predict_next_word(model, input_sequence, N):
    words = preprocess_text(input_sequence)

    if len(words) < N - 1:
        return "Input sequence too short for this N-gram model."

    context = tuple(words[-(N - 1):])

    if context in model:
        next_word_probs = model[context]

        best_next_word = max(next_word_probs, key=next_word_probs.get)

        print(f"  [Math] Probabilities for context {context}: {next_word_probs}")

        return best_next_word
    else:
        return "No prediction available"
```

```python
if __name__ == "__main__":
    print("=== N-GRAM LANGUAGE MODEL EXECUTION ===\n")

    corpus = """Data science is an interdisciplinary field.
    Data science uses statistics and machine learning.
    Big data analytics is important in modern applications.
    Machine learning improves data science solutions."""

    N = 3

    print(" Step 1: Training Phase Started...")
    words = preprocess_text(corpus)
    model = train_ngram_model(words, N)
    print(" Model training completed successfully!\n")

    input_seq = "Data science uses"

    print(" Input Sequence Entered:")
    print(f"   {input_seq}")

    prediction = predict_next_word(model, input_seq, N)

    print("\n Prediction Result:")
    print(f"   Next word prediction => {prediction}")

    print("\n=== PROCESS COMPLETED ===")
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk import pos_tag, RegexpParser
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
def extract_noun_phrases(paragraph):

    np_list = []


    sentences = sent_tokenize(paragraph)


    grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
    chunk_parser = RegexpParser(grammar)

    for sentence in sentences:

        words = word_tokenize(sentence)

        tagged_words = pos_tag(words)

        parse_tree = chunk_parser.parse(tagged_words)

        for subtree in parse_tree.subtrees():
            if subtree.label() == "NP":

                np = " ".join(word for word, tag in subtree.leaves())
                np_list.append(np)


    return np_list
```

```python
if __name__ == "__main__":
    print("--- Starting Noun Phrase Extraction ---")

    text = "Data science methods are essential in modern business analytics. The advanced algorithms process complex data from multiple sources."

    print(f"\nInput Paragraph:\n{text}")
    print("-" * 50)

    extracted_nps = extract_noun_phrases(text)

    print("Extracted Noun Phrases:")
    for np in extracted_nps:
        print(f"-> {np}")

    print("\n--- END ---")
```
